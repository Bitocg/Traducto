#### Importamos las bibliotecas de MarianMT (el modelo y el tokenizador).
##### He seleccionado esta arquitectura ya que fue diseñada para ser eficiente en términos de recursos computacionales, lo que la hace más rápida y menos intensiva en memoria en comparación con algunos otros modelos de traducción automática.

```python
from transformers import MarianMTModel, MarianTokenizer
```

#### Definimos el nombre del modelo y cargamos el modelo y el tokenizador

```python
model_name = "Helsinki-NLP/opus-mt-es-en"
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)
```

#### Creamos la función de inferencia

```python
def translate(text):
    # Tokenizar y convertir el texto a la representación del modelo
    input_ids = tokenizer.encode(text, return_tensors="pt")

    # Obtener la traducción del modelo
    output = model.generate(input_ids)

    # Decodificar la salida del modelo y devolver la traducción
    translation = tokenizer.decode(output[0], skip_special_tokens=True)
    return translation
```

#### Llamamos a la función y vemos como traduce

```python
spanish_text = "Hola buenos días."
english_translation = translate(spanish_text)
print(f"Texto en inglés: {english_translation}")
```
